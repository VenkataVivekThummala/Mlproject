```python
f,ax=plt.subplots(1,2,figsize=(20,10))
sns.countplot(x=df['gender'],data=df,palette ='bright',ax=ax[0],saturation=0.95)
for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=20)
    
plt.pie(x=df['gender'].value_counts(),labels=['Male','Female'],explode=[0,0.1],autopct='%1.1f%%',shadow=True,colors=['#ff4d4d','#ff8000'])
plt.show()
```

```python
gender_group = df.groupby('gender')[['math_score','reading_score','writing_score','Average_score']].mean()
gender_group
```

```python
plt.figure(figsize=(10, 8))

X = ['Total Average','Math Average']


female_scores = [gender_group["Average_score"][0], gender_group['math_score'][0]]
male_scores = [gender_group['Average_score'][1], gender_group['math_score'][1]]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
```

```python
plt.figure(figsize=(20,15))

group_a=[race_group['Average_score'][0],race_group["math_score"][0]]

group_b=[race_group['Average_score'][1],race_group["math_score"][1]]

group_c=[race_group['Average_score'][2],race_group["math_score"][2]]

group_d=[race_group['Average_score'][3],race_group["math_score"][3]]

group_e=[race_group['Average_score'][4],race_group["math_score"][4]]

x=['Group A','Group B','Group C','Group D','Group E']

X_axis=np.arange(len(x))

plt.bar(X_axis+0.2,group_a,0.2,label="Group A")
plt.bar(X_axis-0.2,group_b,0.2,label="Group B")
plt.bar(X_axis+0.2,group_c,0.2,label="Group C")
plt.bar(X_axis-0.2,group_d,0.2,label="Group D")
plt.bar(X_axis+0.2,group_e,0.2,label="Group E")

plt.xticks(X_axis,x)
plt.ylabel("Marks")
plt.title("Scores According to the race/ethnicity",fontweight='bold')
plt.show()

```
